```python
from transformers import AutoModel, AutoTokenizer

model_name = "AIA-Vietnamese-Embedding"  # Thay bằng tên model nếu có

# Tải model và tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Lưu model về máy
model.save_pretrained("./aia_embedding")
tokenizer.save_pretrained("./aia_embedding")

print("Model đã được tải và lưu tại ./aia_embedding")
```
